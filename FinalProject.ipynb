### How did United Airlines’ average departure delays at EWR compare between 2018 and 2023?

##### Here I took the last four characters of each value in the date column, which gives me the year, then used that as the grouping category. Finally I grouped all the rows by those years and calculated the average departure delay for each year.

```python
avg_delay = df.groupby(df['Date (MM/DD/YYYY)'].str[-4:])['Departure delay (Minutes)'].mean()
avg_delay.plot(kind='bar', color='blue')
plt.title("Average Departure Delays (Minutes) for United at EWR")
plt.ylabel("Average Delay (minutes)")
plt.xlabel("Year")
plt.show()
```

##### Summary: The bar chart comparing average departure delays per year provides a direct look at overall delay performance for United flights from EWR. A higher average delay in 2023 would suggests that operational efficiency has declined.

### Are certain times of day more prone to delays for United flights departing EWR?

##### Under scheduled departure time I kept getting this error code, "ValueError: cannot convert float NaN to integer", so I extracted the hour as a float first to handle NaNs, filled missing hours with mean, then grouped by hour and calculated average departure delay.

```python
df['DepHour'] = df['Scheduled_departure_time'].str[:2].astype(float)
df['DepHour'] = df['DepHour'].fillna(df['DepHour'].mean()).astype(int)
hourly_delay = df.groupby('DepHour')['Departure delay (Minutes)'].mean()

hourly_delay.plot(kind='line', marker='.')
plt.title("Average Departure Delay by Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Average Delay (minutes)")
plt.grid(True)
plt.savefig("delay_hour.png")
plt.show()
```

##### Summary: The line chart shows average departure delays by hour of the day, revealing patterns in delays across different scheduled departure times. We can see here that flights are more prone to disruptions after 12pm. This information can help identify times of day that may require improved scheduling, additional staffing, or contingency planning to reduce delays.

### Which delay causes (weather, carrier, NAS, late aircraft) were most significant in each year?

##### First I created a list of the delay related columns, then extracted the year from each date by taking the last four characters of the date string and storing them in a new column called year. Finally, grouped the dataset by year and added up all delay minutes for each cause.

```python
cause_cols = ['Delay Carrier (Minutes)', 'Delay Weather (Minutes)',
              'Delay National Aviation System (Minutes)', 'Delay Late Aircraft Arrival (Minutes)']
df['Year'] = df['Date (MM/DD/YYYY)'].str[-4:]
delay_causes = df.groupby('Year')[cause_cols].mean()
delay_causes.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title("Average Delay Minutes by Cause (2018 vs 2023)")
plt.ylabel("Average Delay (minutes)")
plt.xlabel("Year")
plt.legend(title="Cause")
plt.savefig("delay_cause.png")
plt.show()
```

##### Summary: The stacked bar chart of total delay minutes by cause breaks down the contribution of Carrier, Weather, National Aviation System, and Late Aircraft Arrival delays. Comparing 2018 and 2023 highlights which factors were the most significant contributors to delays in each year. We can conlude here that late aircraft arrivals are the highest cause of delays.

### What percentage of United flights departed on time in 2018 vs 2023?

##### First I created a new column called OnTime that marks each flight as either on time (True) if its departure delay is zero or less, or late (False) if the delay is positive. Then I grouped the flights by year and calculated the average of the OnTime values, which works because True counts as 1 and False counts as 0. This average represents the proportion of flights that were on time, and multiplying by 100 converts that proportion into a percentage for each year.

```python
df['OnTime'] = df['Departure delay (Minutes)'] <= 0
on_time_percentage = df.groupby('Year')['OnTime'].mean() * 100
on_time_percentage.plot(kind='bar', color='royalblue')
plt.title("Percentage of On-Time Departures (2018 vs 2023)")
plt.ylabel("On-Time %")
plt.xlabel("Year")
plt.savefig("ontime_percent.png")
plt.show()
```

##### Summary: The bar chart shows the percentage of on-time departures, illustrating how often United flights from EWR departed as scheduled in 2018 compared to 2023. The decrease in this percentage in 2023 indicates that flights experienced more frequent delays.
